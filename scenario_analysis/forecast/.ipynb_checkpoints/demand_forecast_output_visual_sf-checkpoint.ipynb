```python
import pandas as pd
from pandas import read_csv
import os
from os import listdir
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')

os.chdir('/Users/xiaodanxu/Documents/SynthFirm.nosync/')

plt.style.use('ggplot')
sns.set(font_scale=1.4)  # larger font
```

```python
analysis_years = [2018, 2040, 2050]
aus_faf_code = [481, 488, 489]
combined_b2b_flow = None

for year in analysis_years:
    directory_to_load = 'outputs_aus_' + str(year)
    b2b_flow_by_year = read_csv(directory_to_load + '/processed_b2b_flow_summary.csv')
    b2b_flow_by_year.loc[:, 'Year'] = year
    combined_b2b_flow = pd.concat([combined_b2b_flow, b2b_flow_by_year])
    
combined_b2b_flow.head(5)    
```

```python
# total production
print(len(combined_b2b_flow))
austin_production = \
combined_b2b_flow.loc[combined_b2b_flow['orig_FAFID'].isin(aus_faf_code)]
print(len(austin_production))
austin_production_total = \
austin_production.groupby(['Year', 'mode_choice'])[['ShipmentLoad']].sum()
austin_production_total = austin_production_total.reset_index()
austin_production_to_plot = \
pd.pivot_table(austin_production_total,
               values='ShipmentLoad', 
               index=['Year'],
               columns=['mode_choice'], 
               aggfunc=np.sum, sort = False)
austin_production_to_plot.loc[:, 'Total load'] = austin_production_to_plot.sum(axis = 1)
austin_production_to_plot.loc[:, 'Truck load'] = \
austin_production_to_plot.loc[:, 'For-hire Truck'] + austin_production_to_plot.loc[:, 'Private Truck']
austin_production_to_plot = austin_production_to_plot.reset_index()
austin_production_to_plot = austin_production_to_plot[['Year', 'Total load', 'Truck load']]

austin_production_to_plot.head(5)
```

```python
# total consumption
print(len(combined_b2b_flow))
austin_consumption = \
combined_b2b_flow.loc[combined_b2b_flow['dest_FAFID'].isin(aus_faf_code)]
print(len(austin_production))
austin_consumption_total = \
austin_consumption.groupby(['Year', 'mode_choice'])[['ShipmentLoad']].sum()
austin_consumption_total = austin_consumption_total.reset_index()
austin_consumption_to_plot = \
pd.pivot_table(austin_consumption_total,
               values='ShipmentLoad', 
               index=['Year'],
               columns=['mode_choice'], 
               aggfunc=np.sum, sort = False)
austin_consumption_to_plot.loc[:, 'Total load'] = austin_consumption_to_plot.sum(axis = 1)
austin_consumption_to_plot.loc[:, 'Truck load'] = \
austin_consumption_to_plot.loc[:, 'For-hire Truck'] + austin_consumption_to_plot.loc[:, 'Private Truck']
austin_consumption_to_plot = austin_consumption_to_plot.reset_index()
austin_consumption_to_plot = austin_consumption_to_plot[['Year', 'Total load', 'Truck load']]
austin_consumption_to_plot.head(5)
```

```python
austin_production_to_plot.columns = \
['Year', 'Total production (1000 ton)', 'Production shipped by truck (1000 ton)']
austin_consumption_to_plot.columns = \
['Year', 'Total consumption (1000 ton)', 'Consumption shipped by truck (1000 ton)']
austin_flow_to_plot = pd.merge(austin_production_to_plot,
                              austin_consumption_to_plot,
                              on = 'Year', how = 'left')
austin_flow_to_plot.loc[:, 'Total load'] = \
austin_flow_to_plot.loc[:, 'Total production (1000 ton)'] + \
austin_flow_to_plot.loc[:, 'Total consumption (1000 ton)']

austin_flow_to_plot.loc[:, 'Truck load'] = \
austin_flow_to_plot.loc[:, 'Production shipped by truck (1000 ton)'] + \
austin_flow_to_plot.loc[:, 'Consumption shipped by truck (1000 ton)']
austin_flow_to_plot.head(5)
```

```python
# plt.figure(figsize = (6,5))
austin_flow_to_plot[['Year', 'Total load', \
                     'Truck load']].plot(kind = 'bar', x = 'Year', figsize=(8,6))
plt.xticks(rotation = 60)
plt.ylabel('Shipment load (1000 ton)')
plt.savefig('plots/SynthFirm_demand_forecast.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
austin_flow_to_plot.to_csv('outputs/SynthFirm_demand_forecast_aus.csv')
```
